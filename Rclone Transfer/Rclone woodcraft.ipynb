####<font color=GREEN size=4><b>üîó Gdrive Downloader</b></font>

```python
# @markdown <font color=GREEN size=4><b>üîß Gdrive Downloader</b></font>
# @markdown <p>‚¨§ Change ***mode*** to \"file\" for link from single file link and \"folder\" for link from folder link. Fill ***id_link*** with only the id from url, not full url.</p>
# @markdown <center><img src="https://raw.githubusercontent.com/SudoR2spr/GoFile-Downloader/main/Op%20Source/logo/gdrive.png" alt="GDrive logo" height="50"></center>
import os
# @markdown <font color=GREEN size=4><b>üîß Gdrive MODE</b></font>
mode = "file" # @param ["file", "folder"]
id_link = "" # @param {type:"string"}
output_path = "/content/drive/My Drive/Rclone" # @param {type:"string"}

import os
delete_path = '/content/sample_data'
if os.path.exists(delete_path):
    !rm -rf '/content/sample_data'

try:
    os.makedirs(output_path, exist_ok=True)
except OSError as error:
    pass
os.chdir(output_path)

if mode == "file":
    command = f"gdown --remaining-ok {id_link}"
else:
    command = f"gdown --folder --remaining-ok {id_link}"

!{command}

os.chdir('/content')
```

####<font color=blue size=4><b>üìÅ Rclone Transfer</b></font>

```python
# @markdown <font color=Green size=4><b>üìÅ Transfer Files with Rclone</b></font>
# @markdown <center><img src="https://raw.githubusercontent.com/SudoR2spr/SudoR2spr/main/assets/angel-op/rclone.png" alt="gofile logo" height="50"></center>
# @markdown <p>‚¨§ Use <b>rclone_version</b> stable for official rclone, and <b>rclone_version</b> mod with additional support for teldrive. Fill <b>source</b> with source path or direct download link, <b>dest</b> with your destination path.</p>
rclone_version = "stable" # @param ["stable", "mod"]
source = "" # @param {type:"string"}
dest = "" # @param {type:"string"}

import os, urllib.request

if not os.path.exists("/content/peeklab.cpython-310-x86_64-linux-gnu.so"):
    urllib.request.urlretrieve("https://github.com/SudoR2spr/GoFile-Downloader/raw/main/Op%20Source/peeklab.cpython-310-x86_64-linux-gnu.so", "/content/peeklab.cpython-310-x86_64-linux-gnu.so")

from peeklab import installRclone, installRcloneM, uploadConf
from IPython.display import clear_output

if rclone_version == "stable":
    installRclone()
else:
    installRcloneM()

uploadConf()

if source.startswith("https://"):
    command = f"rclone copyurl -v --stats 1 --stats-one-line --checkers 16 --onedrive-encoding \"Slash\" -a --header-filename \"{source}\" \"{dest}\""
else:
    command = f"rclone copy -v --stats 1 --stats-one-line --checkers 16 --onedrive-encoding \"Slash\" --create-empty-src-dirs \"{source}\" \"{dest}\""

!{command}
```
